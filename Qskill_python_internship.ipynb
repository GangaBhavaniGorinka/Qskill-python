```python
# Beginner Task: Data Analysis & Visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# Step 1: Create Student Dataset
# -------------------------------
data = {
    "Student_ID": list(range(1, 31)),
    "Name": [f"Student_{i}" for i in range(1, 31)],
    "Age": [
        18,19,20,21,22,19,20,21,22,18,
        19,20,21,22,23,18,19,20,21,22,
        18,19,20,21,22,23,19,20,21,22
    ],
    "Marks": [
        65,72,80,75,88,70,82,78,90,68,
        74,85,79,92,87,66,71,83,76,89,
        69,73,81,77,86,91,70,84,78,88
    ]
}

df = pd.DataFrame(data)

# Save CSV file
df.to_csv("student_data_30.csv", index=False)

print("CSV file created successfully!\n")

# -------------------------------
# Step 2: Display Dataset
# -------------------------------
print("First 5 records:")
print(df.head())

# -------------------------------
# Step 3: Data Analysis
# -------------------------------
print("\nAverage Marks:", df["Marks"].mean())
print("Maximum Marks:", df["Marks"].max())
print("Minimum Marks:", df["Marks"].min())

# -------------------------------
# Step 4: Data Visualization
# -------------------------------

# Bar Chart
plt.figure()
plt.bar(df["Name"][:10], df["Marks"][:10])
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Marks of First 10 Students")
plt.show()

# Scatter Plot
plt.figure()
plt.scatter(df["Age"], df["Marks"])
plt.xlabel("Age")
plt.ylabel("Marks")
plt.title("Age vs Marks")
plt.show()

# Heatmap
plt.figure()
sns.heatmap(df[["Age", "Marks"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# -------------------------------
# Step 5: Observation
# -------------------------------
top_student = df.loc[df["Marks"].idxmax()]
print("\nTop Scorer:")
print(top_student)
```
